```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

for i in range(2,7):
    df1 = pd.read_csv(f'EP_k_{i}_Nlat', sep=', ',header=None, engine = 'python')
    df1 = df1.transpose()
    df2 = pd.read_csv(f'EP_k_{i}_Fair', sep=', ',header=None, engine = 'python')
    df2 = df2.transpose()

    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
    axes[0].plot(df1.index, df1[0])
    axes[0].title.set_text(f'Nlat @ k = {i}')
    axes[1].plot(df2.index, df2[0])
    axes[1].title.set_text(f'Fair @ k = {i}')
    fig.tight_layout()
```
